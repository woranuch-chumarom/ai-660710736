#**การจัดการข้อมูลหาย โดยใช้วิธีเติมค่าว่างที่ field ชื่อ ด้วยข้อความ "ไม่ระบุ" แทน**
เหตุผลที่ใช้วิธีนี้เพราะชื่อค่อนข้างอนุมานได้ยากจึงเลือกใช้วิธีนี้แทน

```python
df['ชื่อ'].fillna('ไม่ระบุ', inplace=True)
df
```

#**การจัดการค่าผิดปกติที่ field อายุและGPA โดยค่าที่ผิดปกติจะถูกจัดเป็นค่าว่าง และทำการเติมค่าด้วยค่าเฉลี่ยแทน และการจัดการข้อมูลหาย โดยใช้วิธีเติมค่าด้วยค่าเฉลี่ย**
เหตุผลที่ใช้วิธีนี้เพราะวิธีนี้ช่วยทำความสะอาดและเติมเต็มค่าว่างได้

```python
df.loc[df['อายุ'] > 100, 'อายุ'] = np.nan
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)
df['อายุ'] = df['อายุ'].round().astype(int)
df
```

```python
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
df
```

#**การทำชื่อคณะและเพศให้เป็นมาตรฐาน โดยการ map ชื่อคณะและเพศที่มีหลายแบบ ให้เหลือแค่รูปแบบมาตรฐาน เพื่อให้ข้อมูลเป็นหมวดหมู่เดียวกัน**
เหตุผลที่ใช้วิธีนี้ เพราะ สามารถ map ข้อมูลที่มีความแตกต่างกันเป็นรูปแบบเดียวกันได้

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวะ', 'วิศวฯ', 'Engineering', 'วิศวกรรม']:
        return 'วิศวกรรมศาสตร์'
    else:
        return 'ไม่ระบุ'
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
df
```

```python
def standardize_gender(name):
    if name in ['หญิง', 'ญ', 'Female',  'F']:
        return 'หญิง'
    elif name in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    else:
        return 'ไม่ระบุ'
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_gender)
df
```

#**การจัดการข้อมูลหาย โดยใช้การเติมค่าด้วยค่าเฉลี่ยแทนค่าว่างในส่วนสูง, น้ำหนัก, ชั่วโมงเรียน_ต่อสัปหาห์**
เหตุผลที่ใช้วิธีนี้ เพราะ วิธีนี้เหมาะกับข้อมูลตัวเลขและสามารถหาค่าเฉลี่ยได้

```python
mean_height = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(mean_height).round(1)
df
```

```python
mean_weight = df['น้ำหนัก_kg'].mean()
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].fillna(mean_weight).round(1)
df
```

```python
mean_hr = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hr).round(1)
df
```

#**Feature สร้าง BMI จากส่วนสูงและน้ำหนัก**

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

print(df[['ชื่อ', 'BMI', 'สุขภาพ']])
```
